## **Bucles For**
Los bucles `for` se basan en iteradores que recorren todos los elementos de aquellos objetos que sean iterables: listas, tuplas, cadenas. Ejecuta una pieza de c√≥digo repetidamente en funci√≥n de la cantidad de elementos que contenga el objeto iterable.

Un bucle for establece la variable iteradora en cada valor de una lista, arreglo o cadena proporcionada y repite el c√≥digo en el cuerpo del bucle for para cada valor de la variable iteradora.

En el ejemplo a continuaci√≥n, usamos un bucle for para imprimir cada n√∫mero de nuestro arreglo.

````python
# Ejemplo bucle for
lista =  [1, 2, 3, 4]
for i in lista:
    print(i, end=", ") # prints: 1, 2, 3, 4,
````

# Bucle for en diccionarios
Un caso es especial de bucle for se da al recorrer los elementos de un diccionario. Dado que un diccionario est√° compuesto por pares clave/valor, hay distintas formas de iterar sobre ellas.

1 ‚Äì Recorrer las claves del diccionario.

````python
valores = {'A': 4, 'E': 3, 'I': 1, 'O': 0}
for k in valores:
    print(k)
````

2 ‚Äì Iterar sobre los valores del diccionario

````python
valores = {'A': 4, 'E': 3, 'I': 1, 'O': 0}
for v in valores.values():
    print(v)
````

3 ‚Äì Iterar a la vez sobre la clave y el valor de cada uno de los elementos del diccionario.

````python
valores = {'A': 4, 'E': 3, 'I': 1, 'O': 0}
for k, v in valores.items():
    print('k=', k, ', v=', v)
````

````python
# Iterar sobre los valores de una lista o tupla
A = ["hola", 1, 65, "gracias", [2, 3]]
for value in A:
    print(value)

````

````python
#  Iterar sobre claves en un diccionario (tambi√©n conocido como hashmap)

fruta_a_color = {"manzana": "#ff0000",
                 "lima": "#ffff00",
                 "naranja": "#ffa500"}

for key in fruta_a_color:
    print(key, fruta_a_color[key])
````

Iterar sobre dos listas del mismo tama√±o en un solo bucle con la funci√≥n zip()


````python
A = ["a", "b", "c"]
B = ["a", "d", "e"]

for a, b in zip(A, B):
  print (a, b, a == b)
````

Iterarar sobre una lista y obtener el √≠ndice correspondiente con la funci√≥n enumerate() - enumerar


````python
A = ["esto", "es", "algo", "divertido"]

for indice,palabra in enumerate(A):
    print(indice, palabra)
````

## **Bucles While**
Los bucles o loops while ejecutan los bloques de c√≥digo a los que afecta siempre y cuando la condici√≥n del bucle sea verdadera.

El bucle while se repite hasta que la condici√≥n se ha cumplido un numero determinado de veces.

La sem√°ntica tras esta sentencia es: `**Mientras se cumpla la condici√≥n haz algo**`
*Es similar a la funci√≥n if, sin embargo, el bucle while en Python se ejecuta varias veces mientras que los if tan solo se ejecutan una vez*.

https://www.freecodecamp.org/espanol/news/python-tutorial-ciclos-while/

image.png

Analicemos este proceso en m√°s detalle:

* El proceso inicia cuando se encuentra un ciclo while durante la ejecuci√≥n del programa.
* Se eval√∫a la condici√≥n para determinar si es verdadera o falsa (`True o False`).
* Si la condici√≥n es verdadera (`True`), la secuencia de instrucciones o sentencias que pertenecen al ciclo while se ejecutan.
* Nuevamente se eval√∫a y verifica la condici√≥n del ciclo while.
* Si la condici√≥n es verdadera (`True`) nuevamente, la secuencia de instrucciones en el cuerpo del ciclo while se ejecutan y el proceso se repite.
* Cuando la condici√≥n es falsa (`False`), el ciclo se detiene y el programa contin√∫a m√°s all√° del ciclo while.

Una de las caracter√≠sticas m√°s importantes de los ciclos while es que las variables usadas en la condici√≥n del ciclo no se actualizan autom√°ticamente.

Nosotros tenemos que actualizar sus valores expl√≠citamente en nuestro c√≥digo para asegurarnos de que el ciclo se detendr√° en alg√∫n momento, cuando la condici√≥n sea falsa (`False`).


**Esta es la sintaxis b√°sica:**

image.png

Estos son los elementos principales (en orden):

La palabra clave while (seguida de un espacio).

*   Ua condicni√≥n que determina si el ciclo continuar√° su ejecuci√≥n o no en base a su valor (True o False).
*   Elemento de listaDos puntos (:) al final de la primera l√≠nea.
*   La secuencia de instrucciones o sentencias que se repetir√°n.
*   A este bloque de c√≥digo se le denomina el "cuerpo" del ciclo y debe estar indentado.
*  Si una l√≠nea de c√≥digo no est√° indentada, no se considerar√° parte del ciclo.
