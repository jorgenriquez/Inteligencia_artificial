## Histogramas

La funcin hist permite crear histogramas en Python al utilizar matplotlib. Para crear un histograma bsico tan solo es necesario pasar una variable numrica a la funcin.

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.subplot(211) # 2 rows, 1 columns, 1st subplot
plt.hist(data, bins = 10, rwidth=0.8)

plt.subplot(212) # 2 rows, 1 columns, 2nd subplot
plt.hist(data, bins = [1, 1.5, 2, 2.5, 3], rwidth=0.95)
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

**Histograma horizontal**

Estableciendo el argumento orientation como "horizontal" puedes rotar los ejes y crear un histograma horizontal.

```python
# Semilla para reproducibilidad

np.random.seed(4)

# Simulacin de datos
x = np.random.normal(0, 1, 100)

# Histograma
plt.figure(figsize=(14,6))
fig, ax = plt.subplots()
ax.hist(x, orientation = "horizontal")
plt.show()
```

### Nmero de clases

El nmero por defecto de clases (bins) de un histograma hecho con la funcin `hist` es 10. Sin embargo, esto puede que no sea la mejor opcin para todos los conjuntos de datos. En este escenario, puedes utilizar el argumento `bins` para elegir un nmero de clases ms adecuado o establecer un mtodo de seleccin. Los mtodos de seleccin disponibles son '`auto`', '`fd`', '`doane`', '`scott`', '`stone`', '`rice`', '`sturges`' (el utilizado por defecto en el lenguaje de programacin R) y 'sqrt'.

```python
# Semilla para reproducibilidad
np.random.seed(6)

# Simulacin de datos
x = np.random.normal(0, 1, 500)

# Histograma (demasiadas clases)
fig, ax = plt.subplots()
ax.hist(x, bins = 60)
plt.show()
```

### Histogramas por grupos


Tambin puedes crear un histograma con varios grupos con la funcin. Para ello tan solo es necesario pasar una lista con varios conjuntos de datos, como en el ejemplo siguiente. Por defecto, las barras de los grupos se mostrarn una al lado de las otras.

```python
# Semilla para reproducibilidad
np.random.seed(4)

# Simulacin de datos
x1 = np.random.normal(0, 1, 100)
x2 = np.random.normal(0, 1, 100)
x = [x1, x2]

# Histograma
fig, ax = plt.subplots()
ax.hist(x)
plt.show()
# plt.show()
```

### Histogramas barstacked

El mtodo "`barstacked`" apila las barras cuando hay varios grupos. Este tipo de histograma tambin se puede crear con `stacked = True.`

```python
# Semilla para reproducibilidad
np.random.seed(4)

# Simulacin de datos
x = np.random.normal(0, 1, 100)

# Histograma
fig, ax = plt.subplots()
ax.hist(x, histtype = "step")
# plt.show()
```

### Histograma  stepfilled

El ltimo mtodo disponible es el mtodo "`stepfilled`", que es lo mismo que "`step`" pero con color de relleno. Si tienes un grupo mostrar lo mismo que el tipo por defecto, pero si tienes varios grupos se solaparn uno sobre otro, como en el ejemplo siguiente.

```python
# Semilla para reproducibilidad
np.random.seed(4)

# Simulacin de datos
x1 = np.random.normal(0, 1, 100)
x2 = np.random.normal(0, 1, 100)
x = [x1, x2]

# Histograma
fig, ax = plt.subplots()
ax.hist(x, histtype = "stepfilled")
# plt.show()
```

## Array

```python
"""
======================
streamplot(X, Y, U, V)
======================

See `~matplotlib.axes.Axes.streamplot`.
"""
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make a stream function:
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
# make U and V out of the streamfunction:
V = np.diff(Z[1:, :], axis=1)
U = -np.diff(Z[:, 1:], axis=0)

# plot:
fig, ax = plt.subplots()

ax.streamplot(X[1:, 1:], Y[1:, 1:], U, V)

plt.show()
```

## Trazar series temporales
Aqu trazamos la serie temporal, agregamos etiquetas, ttulo y una pequea anotacin sobre el origen de los datos. Usando la barra de herramientas de la izquierda puede ampliar los datos para ver, por ejemplo, el calentamiento y enfriamiento da/noche. Tambin puede desplazarse a lo largo de la serie temporal utilizando el icono de cuatro flechas. Tenga en cuenta que como hemos dicho sharex=Trueal crear los ejes, la ventana de tiempo permanece conectada en ambos sero de tiempo.

Dataset es tomado de  https://www.ncei.noaa.gov/pub/data/uscrn/products/hourly02/ y es la temperatura horaria y la radiacin solar en Dinosaur, Colorado.

Este ejemplo es tomdao de la pagina oficial de matplotlib: https://mybinder.org/v2/gh/matplotlib/mpl-brochure-binder/main?labpath=MatplotlibExample.ipynb

```python
#   Note the use of datetimes in the file complicate loading a bit.
#   We recommend using pandas or xarray for more elegant solutions
#   to handling complex timeseries data.
with open('/content/small.txt', 'r') as f:
    data = np.genfromtxt(f, dtype='datetime64[s],f,f,f',
                         names=['date', 'doy', 'temp', 'solar'])
datetime = data['date']
dayofyear = data['doy']
temperature = data['temp']
solar = data['solar']

# make two-day smoothed versions:
temp_low = np.convolve(temperature, np.ones(48)/48, mode='same')
solar_low = np.convolve(solar, np.ones(48)/48, mode='same')
```

```python
fig, (ax0, ax1) = plt.subplots(2, 1, sharex=True, constrained_layout=True)

# temperature:
ax0.plot(datetime, temperature, label='hourly')
ax0.plot(datetime, temp_low, label='smoothed')
ax0.legend(loc='upper right')
ax0.set_ylabel('Temperature $[^oC]$')  # note the use of TeX math formatting

# solar-radiation:
ax1.plot(datetime, solar, label='hourly')
ax1.plot(datetime, solar_low, label='smoothed')
ax1.legend(loc='upper right')
ax1.set_ylabel('Solar radiation $[W\,m^{-2}]$')   # note the use of TeX math formatting

ax0.set_title('Observations: Dinosaur, Colorado', loc='left')
ax0.text(0.03, 0.03, 'https://www.ncei.noaa.gov/pub/data/uscrn/products/hourly02/',
         fontsize='small', fontstyle='italic', transform=ax0.transAxes);
```

## Graficos en 3D

```python
"""
==============
3D scatterplot
==============

See `~mpl_toolkits.mplot3d.axes3d.Axes3D.scatter`.
"""
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# Make data
np.random.seed(19680801)
n = 100
rng = np.random.default_rng()
xs = rng.uniform(23, 32, n)
ys = rng.uniform(0, 100, n)
zs = rng.uniform(-50, -25, n)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(xs, ys, zs)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

```python
"""
=====================
3D surface
=====================

See `~mpl_toolkits.mplot3d.axes3d.Axes3D.plot_surface`.
"""
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

plt.style.use('_mpl-gallery')

# Make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

```python
"""
======================
Triangular 3D surfaces
======================

See `~mpl_toolkits.mplot3d.axes3d.Axes3D.plot_trisurf`.
"""
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

plt.style.use('_mpl-gallery')

n_radii = 8
n_angles = 36

# Make radii and angles spaces
radii = np.linspace(0.125, 1.0, n_radii)
angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)[..., np.newaxis]

# Convert polar (radii, angles) coords to cartesian (x, y) coords.
x = np.append(0, (radii*np.cos(angles)).flatten())
y = np.append(0, (radii*np.sin(angles)).flatten())
z = np.sin(-x*y)

# Plot
fig, ax = plt.subplots(subplot_kw={'projection': '3d'})
ax.plot_trisurf(x, y, z, vmin=z.min() * 2, cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

### **Diagrama de caja**


Un diagrama de caja se conoce a veces como diagrama de caja y bigotes. Muestra la distribucin de los datos cuantitativos que representan las comparaciones entre variables. El diagrama de caja muestra los cuartiles del conjunto de datos, mientras que los bigotes se extienden para mostrar el resto de la distribucin, es decir, los puntos que indican la presencia de valores atpicos. Se crea utilizando el mtodo boxplot() .

image.png


Un diagrama de caja es una poderosa herramienta de visualizacin de datos que se utiliza para comprender la distribucin de los datos. Divide los datos en cuartiles y los resume en funcin de cinco nmeros derivados de estos cuartiles:

**mediana:** el valor medio de los datos. marcado como Q2, representa el percentil 50.

**primer cuartil:**  el valor medio entre mnimo no atpico y mediana. marcado como Q1, representa el percentil 25.


**tercer cuartil:** el valor medio entre el mximo no atpico y la mediana. marcado como Q3, representa el percentil 75.


**valor no atpico mximo:**  calculado por (Q3 + 1,5*IQR). Todos los valores por encima de este se consideran valores atpicos.

**valor mnimo no atpico:** calculado por (Q1  1,5*IQR). Todos los valores por debajo de este se consideran valores atpicos.

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1,1, figsize=(8,6))

# loading dataset
data = sns.load_dataset("iris")

sns.boxplot(x='species', y='sepal_width', data=data)
plt.show()
```
