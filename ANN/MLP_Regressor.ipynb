## **Perceptrón multicapa**

Un perceptrón multicapa (**MLP**), también conocido como red neuronal prealimentada (MLP), es un tipo de red neuronal artificial (RNA) utilizada para aprendizaje supervisado. Está compuesta por múltiples capas de neuronas artificiales, donde cada capa está conectada a la siguiente mediante sinapsis con pesos asociados. La capa de entrada recibe los datos de entrada, las capas intermedias procesan la información y la capa de salida genera la predicción o el resultado.

A diferencia del perceptrón simple, que solo tiene una capa de neuronas, el MLP puede modelar relaciones no lineales entre las variables de entrada y salida, lo que le permite abordar una amplia gama de problemas complejos.

image.png

La capa más a la izquierda, conocida como capa de entrada, consta de un conjunto de neuronas. \{`x_i | x_1, x_2, ..., x_m`\}
que representan las características de entrada. Cada neurona de la capa oculta transforma los valores de la capa anterior con una suma lineal ponderada `w_1x_1 + w_2x_2 + ... + w_mx_m`, seguido de una función de activación no lineal
  como la función tan hiperbólica. La capa de salida recibe los valores de la última capa oculta y los transforma en valores de salida.

El módulo contiene los atributos públicos `coefs_ y  intercepts`_. `coefs_` es una lista de matrices de peso, donde la matriz de peso en el índice **`i`**
representa los pesos entre capas **`i`** y capa **`i+1`**. `intercepts_` es una lista de vectores de sesgo, donde el vector en el índice **`i`** representa los valores de sesgo agregados a la capa.

**Las ventajas del Perceptrón multicapa son:**
* Capacidad para aprender modelos no lineales.

* Capacidad para aprender modelos en tiempo real (aprendizaje en línea) usando `partial_fit`.

**Las desventajas del perceptrón multicapa (MLP) incluyen:**


*   Los MLP con capas ocultas tienen una función de pérdida no convexa cuando existe más de un mínimo local. Por lo tanto, diferentes inicializaciones de pesos aleatorios pueden dar lugar a diferentes precisiones de validación.
*   MLP requiere ajustar una serie de hiperparámetros, como la cantidad de neuronas ocultas, capas e iteraciones.

*   MLP es sensible al escalado de funciones.


### **Recomendaciones**

El perceptrón multicapa es sensible al escalado de funciones, por lo que se recomienda encarecidamente escalar sus datos. Por ejemplo, escale cada atributo en el vector de entrada X a [0, 1] o [-1, +1], o estandarícelo para que tenga media 0 y varianza 1. Tenga en cuenta que debe aplicar la misma escala al conjunto de prueba para resultados significativos. Puede utilizarlo StandardScalerpara la estandarización.

### **Hiperparámetros:**

Los hiperparámetros del MLP son configuraciones que definen la estructura y el comportamiento de la red durante el entrenamiento. Algunos de los hiperparámetros más importantes incluyen:

* **Tamaño de las capas ocultas** **`(hidden_layer_sizes)`**: Define el número de neuronas en cada capa oculta **`default`**=(100).
* **Función de activación** **`(activation)`**: `{‘identity’, ‘logistic’, ‘tanh’, ‘relu’}`, default=’relu’ Especifica la función utilizada por las neuronas para introducir no linealidad en el modelo. Algunas opciones comunes incluyen la función sigmoide, la función de activación ReLU (Rectified Linear Unit) y la función tangente hiperbólica, Funciones como ‘***relu***’, ‘***tanh***’, ‘***logistic***’ que determinan la salida de las neuronas.
+ **Optimizador** **`(solver)`**: Determina cómo se actualizan los pesos de la red durante el entrenamiento. Algunos algoritmos populares incluyen el descenso del gradiente estocástico (SGD)  Algoritmos como ‘**`adam`**’, ‘***sgd***’ o ‘***lbfgs***’ para la optimización de pesos.
* **Regularización L2** **`(alpha)`**: Controla la fuerza del término de regularización.
* **Tamaño del lote** **`(batch_size`**): Tamaño de los minibatches para optimizadores estocásticos.
* **Tasa de aprendizaje** **`(learning_rate)`**: Define cómo se actualizan los pesos durante el entrenamiento.
* **Tasa de aprendizaje inicial** **`(learning_rate_init)`**: La tasa de aprendizaje inicial cuando se usan los optimizadores ‘***sgd***’ o ‘***adam***’.
* **Épocas de entrenamiento:** Define el número de veces que el algoritmo de entrenamiento pasa por el conjunto de datos completo durante el entrenamiento.


Algoritmo de optimización: Determina cómo se actualizan los pesos de la red durante el entrenamiento. Algunos algoritmos populares incluyen el descenso del gradiente estocástico (SGD), Adam, RMSprop, entre otros.

## **Red neuronal MLPRegressor**

El MLPRegressor es un algoritmo de regresión basado en redes neuronales artificiales, concretamente en redes neuronales de perceptrón multicapa (MLP). A continuación encontrará una descripción más detallada:

### **Entrenamiento**
En el contexto del Machine Learning, el término `.fit` se refiere al método utilizado para entrenar un modelo de aprendizaje automático con un conjunto de datos. Este proceso implica ajustar los parámetros del modelo en función de los datos de entrenamiento, de manera que el modelo pueda aprender a realizar la tarea deseada.

* **Definición:** El método `.fit` es una función comúnmente disponible en las bibliotecas de Machine Learning, como `scikit-learn` en Python o `TensorFlow`. Se invoca sobre una instancia del modelo ya inicializada, proporcionándole como entrada el conjunto de datos de entrenamiento y, opcionalmente, otras configuraciones de entrenamiento.

*  ** Proceso de entrenamiento: **Durante la ejecución del método `.fit`, el modelo procesa iterativamente los ejemplos del conjunto de entrenamiento. Para cada ejemplo, el modelo realiza una predicción y luego compara esa predicción con el valor real (conocido como etiqueta) asociado al ejemplo. Basándose en esta comparación, el modelo ajusta sus parámetros internos utilizando un algoritmo de optimización, como el descenso del gradiente. El objetivo del entrenamiento es minimizar el error entre las predicciones del modelo y las etiquetas reales, lo que indica que el modelo ha aprendido a realizar la tarea de manera efectiva

### **Predicción**
La "predicción del modelo" se refiere al proceso de utilizar un modelo de aprendizaje automático entrenado para hacer predicciones sobre nuevos datos o datos no vistos. En el contexto del aprendizaje supervisado, donde se entrena un modelo utilizando un conjunto de datos etiquetado que contiene pares de características y etiquetas objetivo, la predicción implica utilizar el modelo para predecir la etiqueta objetivo correspondiente a nuevas instancias de características.

**El proceso de predicción generalmente implica los siguientes pasos:**

*   **Aplicación del modelo:**S e utiliza el modelo de aprendizaje automático entrenado para predecir las etiquetas objetivo para las nuevas instancias de características.
*   **Obtención de las predicciones:** Se obtienen las predicciones del modelo, que pueden ser en forma de valores continuos (en el caso de problemas de regresión) o clases discretas (en el caso de problemas de clasificación).
*   **Evaluación de las predicciones:** Si se dispone de etiquetas verdaderas para las nuevas instancias, se pueden comparar las predicciones del modelo con estas etiquetas para evaluar el rendimiento del modelo en los datos no vistos.

```python
# Implementación de MLPRegressor Hiperparametros
mlp = MLPRegressor(solver='adam', alpha=1e-4, hidden_layer_sizes=(100), activation = 'relu', max_iter=10000, tol=1e-4, random_state=1)

mlp.fit(X_train_scaled, y_train)
#mlp.fit(X_train, y_train)
predict_train = mlp.predict(X_train_scaled)
#predict_train = mlp.predict(X_train)

# prediction
predict_test = mlp.predict(X_test_scaled)
#predict_test = mlp.predict(X_test)
```

## **Metricas**

*   **Error cuadrático medio (Mean Squared Error - MSE):** Es la media de los cuadrados de los errores entre las predicciones del modelo y los valores reales. Cuanto menor sea el valor de MSE, mejor será el rendimiento del modelo.
MSE.


*   **Coeficiente de determinación R^2**   También conocido como el coeficiente de determinación, es una medida de la proporción de la varianza en la variable objetivo que es predecible a partir de las características del modelo. Puede tomar valores entre 0 y 1, donde 1 indica una predicción perfecta y 0 indica que el modelo no es mejor que predecir la media de los valores de la variable objetivo

*   **Raíz del error cuadrático medio (Root Mean Squared Error - RMSE**): Es la raíz cuadrada del MSE y proporciona una medida de la magnitud promedio de los errores en las unidades originales de la variable objetivo. Al igual que MSE, un valor más bajo de RMSE indica un mejor rendimiento del modelo

```python
# Metrics
from sklearn.metrics import mean_squared_error, r2_score

#r2_mlp =r2_score(y_test, predict_test)
r2_mlp =r2_score(y_test, predict_test)
print("R^2 test mlp: %.4f " %r2_mlp)
#mse_mlp = mean_squared_error(y_test,predict_test)
mse_mlp = mean_squared_error(y_test,predict_test)
print("MSE mlp: %.4f" % mse_mlp)
rmse_mlp = np.sqrt(mse_mlp)
print("RMSE mlp: %.4f" %rmse_mlp)
```
