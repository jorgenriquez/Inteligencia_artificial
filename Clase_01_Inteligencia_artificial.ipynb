### Arrays
Un **array** es una estructura de datos que consiste en una coleccin de elementos (valores o variables), cada uno identificado por al menos un ndice o clave. Un array se almacena de modo que la posicin de cada elemento se pueda calcular a partir de su tupla de ndice mediante una frmula matemtica. El tipo ms simple de array es un array lineal, tambin llamado array unidimensional.

En numpy:
* Cada dimensin se denomina **axis**
* El nmero de dimensiones se denomina **rank**
* La lista de dimensiones con su correspondiente longitud se denomina **shape**
* El nmero total de elementos (multiplicacin de la longitud de las dimensiones) se denomina **size**

```python
# Array cuyos valores son todos 0
a = np.zeros((2, 4))
```

```python
a
```

_**a**_ es un array:
* Con dos **axis**, el primero de longitud 2 y el segundo de longitud 4
* Con un **rank** igual a 2
* Con un **shape** igual (2, 4)
* Con un **size** igual a 8

```python
a.shape
```

```python
a.ndim
```

```python
a.size
```

**Evaluacin de expresiones sobre un DataFrame**

```python
# Creacin de un DataFrame inicializndolo con un diccionario de objetios Series
personas = {
    "peso": pd.Series([84, 90, 56, 64], ["Santiago","Pedro", "Ana", "Julia"]),
    "altura": pd.Series({"Santiago": 187, "Pedro": 178, "Julia": 170, "Ana": 165}),
    "hijos": pd.Series([2, 3], ["Pedro", "Julia"])
}

df = pd.DataFrame(personas)
df
```

```python
# Evaluar una funcin sobre una columna del DataFrame
df.eval("altura / 2")
```

```python
# Asignar el valor resultante como una nueva columna
df.eval("media_altura = altura / 2", inplace=True)
df
```

```python
# Aplicar una funcin externa a una columna del DataFrame
def func(x):
    return x + 2

df["peso"].apply(func)
```
